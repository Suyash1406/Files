```python
df = pd.read_csv(r'pract3.csv')
df.head
```

```python
df.age_group.unique() #returns unique age grps

# aggregate functions
df.groupby(df.age_group).min()
df.groupby(df.age_group).max()
df.groupby(df.age_group).mean()
df.groupby(df.age_group).median()
df.groupby(df.age_group).std()
df.groupby(df.age_group).describe()
```

```python
import pandas as pd

# Create a sample dataset
data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45, 25, 30, 35, 40, 45],
    'Income': [50000, 60000, 70000, 80000, 90000, 55000, 65000, 75000, 85000, 95000],
    'Category': ['A', 'A', 'A', 'B', 'B', 'A', 'B', 'B', 'A', 'B']
})

# Group the data by 'Category' and compute summary statistics for 'Income'
summary_stats = data.groupby('Category')['Income'].describe()

# Print the summary statistics
print(summary_stats)
```

```python
import pandas as pd

# Read the Iris dataset into a dataframe
data = pd.read_csv('Iris.csv')

# Group the data by 'species' column
grouped_data = data.groupby('Species')

# Compute the basic statistical details for each species
species_stats = grouped_data.describe()

# Print the statistical details for 'Iris-setosa', 'Iris-versicolor', and 'Iris-virginica'
print("Iris-setosa:")
print(species_stats.loc['Iris-setosa'])
print()

print("Iris-versicolor:")
print(species_stats.loc['Iris-versicolor'])
print()

print("Iris-virginica:")
print(species_stats.loc['Iris-virginica'])

```
