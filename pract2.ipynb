```python
# outlier detection using z-score
# zscore=(data_point-mean)/std.deviation
from scipy import stats
z = np.abs(stats.zscore(df['math_score']))
print(z)
threshold = 1
sample_outlier = np.where(z>threshold)
sample_outlier
```

```python
# outlier removing techniques
# 1)trimming or removing
# 2)quantile based flooring and capping
# 3)mean median mide

# 1 trimming
new_df=df
for i in sample_outlier:
    new_df.drop(i,inplace=True);
new_df
```
